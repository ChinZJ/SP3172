```python
from MiniSim import MiniSpecies as Species
from MiniSim import MiniBoard as Board
```

```python
BOARD_SIZE = 50
TIMESTEPS = 200
STARTING_SEED_COUNT = 3
TIME_SEED_TO_ADULT = 10
PERCENTAGE_SEED_TO_ADULT = 0.02
AVERAGE_TIME_ALIVE_AS_ADULT = 20
PERCENTAGE_ALIVE_BY_AVERAGE_TIME = 0.5
NUMBER_SEEDS_SPAWNED_PER_TIME = 10

simSpecies = Species(
    TIME_SEED_TO_ADULT,
    PERCENTAGE_SEED_TO_ADULT,
    AVERAGE_TIME_ALIVE_AS_ADULT,
    PERCENTAGE_ALIVE_BY_AVERAGE_TIME,
    NUMBER_SEEDS_SPAWNED_PER_TIME
)

simBoard = Board(
    BOARD_SIZE,
    TIMESTEPS,
    STARTING_SEED_COUNT,
    simSpecies
)

simBoard.checkseeds()
simBoard.run()
simBoard.checkseeds()
simBoard.visboard()
```
